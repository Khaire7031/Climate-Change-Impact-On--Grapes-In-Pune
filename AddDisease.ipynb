```python
data['Dengue'] = 0
```

```python

humidity_condition = (data['humidity'] >= 10) & (data['humidity'] <= 80)
# precipitation_condition = (data['precipMM'] >= 0) & (data['precipMM'] <= 0) 
precipitation_condition = data['precipMM'] >=0.01
temperature_condition = (data['tempC'] >= 0) & (data['tempC'] <= 25) 


data['Dengue'] = ( precipitation_condition & humidity_condition).astype(int)
# data['HeatStroke'] = (humidity_condition & precipitation_condition & temperature_condition).astype(int)

```

```python
 data['Dengue'].value_counts()
```

```python

newData = data.head(40000)
# plt.figure(figsize=(10, 6))
plt.plot(newData['Dengue'], color='red', alpha=0.5)
plt.plot(newData['tempC'], color='green', alpha=0.5)
plt.plot(newData['precipMM'], color='black', alpha=0.5)
plt.plot(newData['humidity'], color='yellow', alpha=0.5)

```

```python


# Extracting columns for plotting
x = data['humidity']
y = data['precipMM']
z = data['tempC']
heat_stroke = data['Dengue']

# Plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot with different colors based on HeatStroke
ax.scatter(x[heat_stroke == 0], y[heat_stroke == 0], z[heat_stroke == 0], c='b', marker='o', label='No Common cold')
ax.scatter(x[heat_stroke == 1], y[heat_stroke == 1], z[heat_stroke == 1], c='r', marker='o', label='Common cold')

# Labels
ax.set_xlabel('Humidity')
ax.set_ylabel('Precipitation (mm)')
ax.set_zlabel('Temperature (Â°C)')

# Legend
ax.legend()

plt.show()
```
